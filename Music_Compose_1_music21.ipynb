```python
import music21
from music21 import converter, note, chord
```

```python
dataset_name = 'cello'
filename = 'cs1-1pre'
file = "./data/{}/{}.mid".format(dataset_name, filename)

original_score = converter.parse(file).chordify()
# chordify 메서드를 사용하여 여러 파트에서 동시에 연주되는 화음으로 압축
# converter.parse :Given a file path, encoded data in a Python string, or a URL, attempt to parse the item into a Stream. 

# Create a chordal reduction of polyphonic music, where each change to a new pitch results in a new chord.
# If a Score or Part of Measures is provided, a Stream of Measures will be returned.
```

```python
original_score.show()
# Displays an object in a format provided by the fmt argument or, if not provided, 
# the format set in the user’s Environment
```

```python
original_score.show('text')
```

```python
# 코드와 악보를 순회하며 각 음표와 쉼표의 피치와 박자(duration)을 두 개의 리스트로 추출, 코드 전체는 하나의 문자열로 저장되고 코드의 개별 음표는
# 점으로 구분. 음표 뒤의 숫자는 음표가 속한 옥타브(octave)를 지칭한다.
notes = []
durations = []

for element in original_score.flat:
    # stream.flat:
    # A very important read-only property that returns a new Stream that has all sub-containers “flattened” within
    # it, that is, it returns a new Stream where no elements nest within other elements.    
    
    if isinstance(element, chord.Chord):
        notes.append('.'.join(n.nameWithOctave for n in element.pitches))
        # pitch.nameWithOctave : Return or set the pitch name with an octave designation. 
        # If no octave as been set, no octave value is returned.
        durations.append(element.duration.quarterLength)
        # duration.quarterLength : Returns the quarter note length or Sets the quarter note length to the
        # specified value. May be expressed as a float or Fraction.

    if isinstance(element, note.Note):
        if element.isRest:
            notes.append(str(element.name))
            # isRest : Boolean read-only value describing if this Note is a Rest (False).
            durations.append(element.duration.quarterLength)
        else:
            notes.append(str(element.nameWithOctave))
            durations.append(element.duration.quarterLength)
            

print('\nduration', 'pitch')
for n,d in zip(notes,durations):
    print(d, '\t', n)
```
